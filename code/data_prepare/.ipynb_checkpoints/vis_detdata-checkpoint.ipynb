```python
from pycocotools.coco import COCO
import json
import numpy as np
import skimage.io as io
import os
import matplotlib.pyplot as plt
import cv2
import pandas as pd
import matplotlib.patches as patches
```

vis coco data

```python
# file paths
imgRoot = "../../user_data/tmp_data/detection/image_tiles/"
annFile = '../../user_data/tmp_data/detection/annotations/round2_tiles_trainval.json'
# load annotations
coco=COCO(annFile)
```

```python
# select images
imgId = 20
img = coco.loadImgs(imgId)[0]
I = cv2.imread(os.path.join(imgRoot, img['file_name']))
```

```python
# visualization
plt.figure(figsize=(10,10))
plt.imshow(I)
annIds = coco.getAnnIds(imgIds=img['id'], catIds=[1], iscrowd=None)
anns = coco.loadAnns(annIds)
coco.showAnns(anns)
```

vis yolov5 data

```python
yolo_imgRoot = "../../user_data/tmp_data/detection/yolov5_data/fbody/images/train/"
yolo_annRoot = "../../user_data/tmp_data/detection/yolov5_data/fbody/labels/train/"
imgs = os.listdir(yolo_imgRoot)
```

```python
img_path = os.path.join(yolo_imgRoot, imgs[40])
ann_path = os.path.join(yolo_annRoot, imgs[40].replace(".jpg", ".txt"))
```

```python
img = cv2.imread(img_path)
anns = pd.read_csv(ann_path, sep=" ", header=None)
img_h, img_w, _ = img.shape
fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(img)
for ann in anns.values:
    cat, xc, yc, w, h = ann
    x,y = int((xc-w/2)*img_w), int((yc-h/2)*img_h)
    w = int(w * img_w)
    h = int(h * img_h)    
    rect = patches.Rectangle((x,y),w,h,linewidth=1, edgecolor='r', facecolor='none')
    ax.add_patch(rect)
plt.show()
```
